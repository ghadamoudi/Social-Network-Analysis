```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Configure plotting in Jupyter
plt.rcParams.update({
    'figure.figsize': (7.5, 7.5),
    'axes.spines.right': False,
    'axes.spines.left': False,
    'axes.spines.top': False,
    'axes.spines.bottom': False})
# Seed random number generator
import random
from numpy import random as nprand
seed = hash("Network Science in Python") % 2**32
nprand.seed(seed)
random.seed(seed)
```

## 1. Simulating Power Law

```python
# Generate the graph
n = 50
p = 0.2
G_erdos = nx.erdos_renyi_graph(n,p, seed =100)# Plot the graph
nx.draw(G_erdos, node_size=10)
```

```python
degree_freq = np.array(nx.degree_histogram(G_erdos)).astype('float')
plt.stem(degree_freq)
plt.ylabel("Frequence")
plt.xlabel("Degree")
plt.show()
```

```python
# Generate the graph
n = 1000000
m = 1
G_barabasi = nx.barabasi_albert_graph(n,m)
```

```python
degree_freq = np.array(nx.degree_histogram(G_barabasi)).astype('float')
plt.stem(degree_freq)
plt.ylabel("Frequence")
plt.xlabel("Degree")
plt.show()
```

```python
plt.plot(degree_freq, 'o')
plt.ylabel("Frequence")
plt.xlabel("Degree")
plt.xscale('log')
plt.yscale('log')
plt.show()
```
