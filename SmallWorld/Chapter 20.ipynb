# Small World Phenomenon

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# n: number of nodes
# m: Each node is connected to k nearest neighbors in ring topology
# p: The probability of rewiring each edge

n = 10 
k = 4
p = 0.5
```

```python
G = nx.watts_strogatz_graph(n, k, p)
pos = nx.circular_layout(G) 
```

```python
plt.figure(figsize = (8, 8))
nx.draw_networkx(G, pos)
```

```python
nx.average_shortest_path_length(G), nx.average_clustering(G)
```

Relatively small average shortest path, and high clustering coefficient, so we have a network that satisfy the small world phenomenon.
